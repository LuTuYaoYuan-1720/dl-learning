```python
# 语义分割要对图像中的每个像素分类，全卷积网络去掉了以前卷积网络最后的全局池化层和全连接层，
# 替换成了转置卷积层，输出的通道就是像素的分类个数，大小为图像大小
```

```python
%matplotlib inline
import torch
import torchvision
from torch import nn
from torch.nn import functional as F
from d2l import torch as d2l

import os
os.environ["KMP_DUPLICATE_LIB_OK"] = "TRUE"
```

```python
# 全卷积先用卷积网络抽取图像特征，用1✕1的卷积层将通道数变换为类别个数
# 再通过转置卷积将高宽变换为输入图像的尺寸
pretrained_net = torchvision.models.resnet18(pretrained=True)
list(pretrained_net.children())[-3:]
```

```python
net = nn.Sequential(*list(pretrained_net.children())[:-2])
```

```python
# 前边的卷积层将图像变换成 512 通道 10 * 15 大小
X = torch.rand(1, 3, 320, 480)
net(X).shape
```

<img src="./pic/转置卷积形状.png" style="zoom=70%"/>

```python
num_classes = 21
net.add_module('final_conv', nn.Conv2d(512, num_classes, kernel_size=1))
net.add_module('transpose_conv', nn.ConvTranspose2d(num_classes, num_classes, 
                                                    kernel_size=64, padding=16, stride=32))
```

```python
# 初始化转置卷积层
# 将图像放大（上采样），双线性插值比较常用，可用于初始化转置卷积层
def bilinear_kernel(in_channels, out_channels, kernel_size):
    factor = (kernel_size + 1) // 2
    if kernel_size % 2 == 1:
        center = factor - 1
    else:
        center = factor - 0.5
    og = (torch.arange(kernel_size).reshape(-1, 1),
          torch.arange(kernel_size).reshape(1, -1))
    filt = (1 - torch.abs(og[0] - center) / factor) * \
           (1 - torch.abs(og[1] - center) / factor)
    weight = torch.zeros((in_channels, out_channels,
                          kernel_size, kernel_size))
    weight[range(in_channels), range(out_channels), :, :] = filt
    return weight
```

```python
# 一个将高宽放大两倍的转置卷积层
conv_trans = nn.ConvTranspose2d(3, 3, kernel_size=4, padding=1, stride=2, bias=False)
conv_trans.weight.data.copy_(bilinear_kernel(3, 3, 4))
```

```python
img = torchvision.transforms.ToTensor()(d2l.Image.open('../data/img/catdog.jpg'))
# img.shape = torch.Size([3, 561, 728])
X = img.unsqueeze(0)
Y = conv_trans(X)
```

```python
out_img = Y[0].permute(1, 2 ,0).detach()
d2l.set_figsize()
d2l.plt.imshow(img.permute(1, 2, 0))
```

```python
# 使用双线性插值法初始化转置卷积层，
W = bilinear_kernel(num_classes, num_classes, 64)
net.transpose_conv.weight.data.copy_(W);
```

```python
batch_size, crop_size = 32, (320, 480)
train_iter, test_iter = d2l.load_data_voc(batch_size, crop_size)
```

```python
# # 对单张图片进行预测的结果是 [1, 21, 320, 480] ,和标签[1, 320, 480] 计算损失
# # 先在通道维上计算损失，21个通道代表在每个类上的预测值
# for imgs, labels in train_iter:
#     pred = net(imgs[0].unsqueeze(0))
#     # img 1 的损失
#     l = F.cross_entropy(pred, labels[0].unsqueeze(0), reduction='none')
#     print(l, l.shape)
#     # img 1 在 [0, 0] 像素上的损失
#     print(F.cross_entropy(pred[0, :, 0, 0].reshape(1, 21), torch.tensor(labels[0, 0, 0]).reshape(1)))
#     break
```

```python
# input (batch_size, 3, 320, 480)  target (batch_size, 320, 480)
def loss(inputs, target):
    # 前边算出来的是在每一个像素上的损失，一张照片在每一个像素上的损失构成了一个[320, 480] 然后取平均
    return F.cross_entropy(inputs, target, reduction='none').mean(1).mean(1)
```

```python
num_epochs, lr, wd, devices = 5, 0.001, 1e-3, d2l.try_all_gpus()
trainer = torch.optim.SGD(net.parameters(), lr=lr, weight_decay=wd)
# d2l.train_ch13(net, train_iter, test_iter, loss, trainer, num_epochs, devices)
```

```python
# 加载模型
net.load_state_dict(torch.load('./voc_net.params'))
net.eval()
net.to(devices[0])
```

```python
# 预测时， 要将输入图像在通道上做标准化，转成四维输入格式
def predict(img):
    X = test_iter.dataset.normalize_image(img).unsqueeze(0)
    # 在21个通道上找到最大的索引
    pred = net(X.to(devices[0])).argmax(dim=1)
    return pred.reshape(pred.shape[1], pred.shape[2])
```

```python
# 预测出来的是每一个像素点的类别， 将其转化为 RGB 值
def label2image(pred):
    colormap = torch.tensor(d2l.VOC_COLORMAP, device=devices[0])
    X = pred.long()
    return colormap[X]
```

```python
voc_dir = '../data/VOCdevkit/VOC2012/'
test_images, test_labels = d2l.read_voc_images(voc_dir, is_train=False)
```

```python
n, imgs = 5, []
for i in range(n):
    crop_rect = [0, 0, 320, 480]
    X = torchvision.transforms.functional.crop(test_images[i], *crop_rect)
    pred = label2image(predict(X))
    imgs += [X.permute(1, 2, 0),
             pred.cpu(),
             torchvision.transforms.functional.crop(test_labels[i], *crop_rect).permute(1, 2, 0)]
d2l.show_images(imgs[0::3] + imgs[1::3] + imgs[2::3], 3, n, scale=1.5)
```
