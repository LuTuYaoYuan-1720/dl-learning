```python
import torch
from torch import nn
from d2l import torch as d2l
```

<img src="./pic/AlexNet.png" style="zoom:70%" />

```python
net = nn.Sequential(
    nn.Conv2d(1, 96, kernel_size=11, stride=4, padding=1), nn.ReLU(),
    nn.MaxPool2d(kernel_size=3, stride=2),
    nn.Conv2d(96, 256, kernel_size=5, padding=2), nn.ReLU(),
    nn.MaxPool2d(kernel_size=3, stride=2),
    nn.Conv2d(256, 384, kernel_size=3, padding=1), nn.ReLU(),
    nn.Conv2d(384, 384, kernel_size=3, padding=1), nn.ReLU(),
    nn.Conv2d(384, 256, kernel_size=3, padding=1), nn.ReLU(),
    nn.MaxPool2d(kernel_size=3, stride=2),
    nn.Flatten(),
    nn.Linear(6400, 4096), nn.ReLU(),
    nn.Dropout(p=0.5),
    nn.Linear(4096, 4096), nn.ReLU(),
    nn.Dropout(p=0.5),
    nn.Linear(4096, 10)
)
```

```python
X = torch.rand(1, 1, 224, 224)
for layer in net:
    X = layer(X)
    print(layer.__class__.__name__, 'output shape: \t', X.shape)
```

```python
batch_size = 64
train_iter, test_iter = d2l.load_data_fashion_mnist(batch_size=batch_size, resize=224)
```

```python
lr, num_epochs = 0.1, 10
d2l.train_ch6(net, train_iter, test_iter, num_epochs, lr, d2l.try_gpu())
```

```python
# batchsize = 64, lr = 0.1  acc 大约92
# batchsize = 32, lr = 0.1  acc 大约92  感觉有点过拟合
# batchsize = 32, lr = 0.15  epoch=9时突然acc掉下来
# batchsize = 64, lr = 0.15  acc 大约92
# batchsize = 128, lr = 0.15  acc 大约90
# batchsize = 256, lr = 0.15 train acc 0.909, test acc 0.901
# batchsize = 256, lr = 0.1 train acc 0.905, test acc 0.895
d2l.evaluate_accuracy_gpu(net, test_iter)
```
